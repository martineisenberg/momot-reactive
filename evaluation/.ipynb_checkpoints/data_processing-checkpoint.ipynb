```python
import seaborn as sns
import pandas as pd
import scipy.stats as st
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
import warnings
from statistics import mean
import ast

warnings.filterwarnings('ignore')
```

# RQ1/2: Measurement of Quality 

```python
files = os.listdir(os.path.join('.', 'quality'))
mean = pd.DataFrame(columns = ["Option", "Naive Repair", "Replanning", "Replanning (seed 0.1)"])
std = pd.DataFrame(columns = ["Option", "Naive Repair", "Replanning", "Replanning (seed 0.1)"])
for f in files:
    if f.endswith("csv"):
        df = pd.read_csv(os.path.join('.', 'quality', f))
        mean_row = {}
        std_row = {}
        for name, group in df.groupby(by='variant'):
            #print(f"\tVariant: {name}")
            #print(f"\t\tFinal objective (mean): {group['final_objective_value'].mean():.3f} ({group['final_objective_value'].std():.3f})\n")
            if(name == "NSGAII_NaivePlanningStrategy_2"):
                mean_row["Naive Repair"] = round(group['final_objective_value'].mean(), 3)
                std_row["Naive Repair"] = round(group['final_objective_value'].std(), 3)
            elif(name == "NSGAII_EvaluationReplanningStrategy-10000_1"):
                mean_row["Replanning"] = round(group['final_objective_value'].mean(), 3)
                std_row["Replanning"] = round(group['final_objective_value'].std(), 3)
            elif(name == "NSGAII_EvaluationReplanningStrategy-10000reusePortion=0.2_withReinit_0"):
                mean_row["Replanning (seed 0.1)"] = round(group['final_objective_value'].mean(), 3)
                std_row["Replanning (seed 0.1)"] = round(group['final_objective_value'].std(), 3)
           
        #print(f)
       # print(mean_row)
        mean_row["Option"] = f.split("@")[1].split(".")[0]
        mean = mean.append(mean_row, ignore_index = True)
        std_row["Option"] = f.split("@")[1].split(".")[0]
        std = std.append(std_row, ignore_index = True)
            # todo: median
            # todo: variance
print("Mean - Absolute Values")
display(mean)

mean["Replanning"] = (mean["Replanning"] - mean["Naive Repair"]) / mean["Naive Repair"] * 100
mean.round({'Replanning': 1})
mean["Replanning (seed 0.1)"] = (mean["Replanning (seed 0.1)"] - mean["Naive Repair"])/ mean["Naive Repair"] * 100
mean.round(3)
print("Mean - Values relative to Naive Repair (in %)")
display(mean)

print("Standard Deviation - Absolut Values")
display(std)
#print(std)
# todo: create table for each rq
# todo: create plot for each rq

# todo: in den CSVs: naive Planning -> naive Repair

```

# RQ2: Measurement of Evaluations/Time
### Predefined standard deviation ( >= 50% decrease from initial std); Search until quality is sufficient (for initial plan & when replanning after disturbance)


```python
files = os.listdir(os.path.join('.', 'threshold_rq1_rq2'))
mean = pd.DataFrame(columns = ["Option", "Replanning", "Replanning (seed 0.2)"])
std = pd.DataFrame(columns = ["Option", "Replanning", "Replanning (seed 0.2)"])
for f in files:
    if f.endswith("csv"):
        df = pd.read_csv(os.path.join('.', 'threshold_rq1_rq2', f))
        mean_row = {}
        std_row = {}
        for name, group in df.groupby(by='variant'):
            #print(f"\tVariant: {name}")
            #print(f"\t\tFinal objective (mean): {group['final_objective_value'].mean():.3f} ({group['final_objective_value'].std():.3f})\n")
            if(name == "NSGAII_ConditionReplanningStrategy-_0-13.6795_0"):
                mean_row["Replanning"] = round(group['initial_planning_time'].mean(), 3)
                std_row["Replanning"] = round(group['initial_planning_time'].std(), 3)
            elif(name == "NSGAII_ConditionReplanningStrategy-_0-13.6795reusePortion=0.2_withSeed_1"):
                mean_row["Replanning (seed 0.2)"] = round(group['initial_planning_time'].mean(), 3)
                std_row["Replanning (seed 0.2)"] = round(group['initial_planning_time'].std(), 3)
           
        #print(f)
       # print(mean_row)
        mean_row["Option"] = f.split("@")[1].split("_errPerDisturbance")[0]
        mean = mean.append(mean_row, ignore_index = True)
        std_row["Option"] = f.split("@")[1].split(".")[0]
        std = std.append(std_row, ignore_index = True)
            # todo: median
            # todo: variance
print("Mean - Replanning time")
display(mean)

#mean["Replanning"] = (mean["Replanning"] - mean["Naive Repair"]) / mean["Naive Repair"] * 100
#mean.round({'Replanning': 1})
mean["Replanning (seed 0.2)"] = (mean["Replanning (seed 0.2)"] - mean["Replanning"])/ mean["Replanning"] * 100
mean.round(3)
print("Mean - Replanning time (with seed) relative to plain replanning (in %)")
display(mean)

print("Standard Deviation - Replanning time")
display(std)
#print(std)
# todo: create table for each rq
# todo: create plot for each rq

# todo: in den CSVs: naive Planning -> naive Repair

```

# RQ3: Parallel Replanning
### Predefined standard deviation ( >= 60% decrease from initial std + initialization approximate); Search until quality is sufficient (initial plan & replan after disturbance while system idle), search in parallel until Q1 = Q2 and T1 < T2

```python
files = os.listdir(os.path.join('.', 'parallel_rq3'))
replan_time = pd.DataFrame(columns = ["Option", "Replanning", "Parallel Rep. 2", "Parallel Rep. 5", "Parallel Rep. 10"])
execution_steps = pd.DataFrame(columns =  ["Option", "Replanning", "Parallel Rep. 2", "Parallel Rep. 5", "Parallel Rep. 10"])
for f in files:
    if f.endswith("csv"):
        df = pd.read_csv(os.path.join('.', 'parallel_rq3', f))
        df_p = pd.read_csv(os.path.join('.', 'parallel_rq3', "parallel_results", f[:-4] + "_predictive.csv"))
        mean_replantime_row = {}
        mean_execsteps_row = {}

        for name, group in df.groupby(by='variant'):
            #print(f"\tVariant: {name}")
            #print(f"\t\tFinal objective (mean): {group['final_objective_value'].mean():.3f} ({group['final_objective_value'].std():.3f})\n")
            if(name == "NSGAII_ConditionReplanningStrategy-_0-10.9436_0"):
                mean_replantime_row["Replanning"] = round(group['last_replanning_time'].mean(), 3)

        for name, group in df_p.groupby(by=['variant', 'predicted_steps']):
            #print(f"\tVariant: {name}")
            #print(f"\t\tFinal objective (mean): {group['final_objective_value'].mean():.3f} ({group['final_objective_value'].std():.3f})\n")
            mean_replantime_row[f"Parallel Rep. {name[1]}"] = round(group['pred_plan_runtime'].mean(), 3)
            mean_execsteps_row["Replanning"] = round(mean([ast.literal_eval(g)[0] for g in group['idle_computed_objectives']]), 3)
            mean_execsteps_row[f"Parallel Rep. {name[1]}"] = round(mean([ast.literal_eval(g)[1]+name[1] for g in group['pred_plan_objectives']]), 3)

        
        #print(f)
       # print(mean_row)
        mean_replantime_row["Option"] = f.split("@")[1].split("_errPerDisturbance")[0]
        mean_execsteps_row["Option"] = f.split("@")[1].split("_errPerDisturbance")[0]
        replan_time = replan_time.append(mean_replantime_row, ignore_index = True)
        execution_steps = execution_steps.append(mean_execsteps_row, ignore_index = True)
        #std_row["Option"] = f.split("@")[1].split(".")[0]
        #std = std.append(std_row, ignore_index = True)
            # todo: median
            # todo: variance
print("Mean - Replanning time (in seconds)")
display(replan_time)

#mean["Replanning"] = (mean["Replanning"] - mean["Naive Repair"]) / mean["Naive Repair"] * 100
#mean.round({'Replanning': 1})
#mean["Replanning (seed 0.2)"] = (mean["Replanning (seed 0.2)"] - mean["Replanning"])/ mean["Replanning"] * 100
#mean.round(3)
#print("Mean - Replanning time (with seed) relative to plain replanning (in %)")
print("Mean - Planned steps after replanning (remaining 2/5/10 steps are included, respectively)")
display(execution_steps)

#print("Standard Deviation - Replanning time")
#display(std)
#print(std)
# todo: create table for each rq
# todo: create plot for each rq

# todo: in den CSVs: naive Planning -> naive Repair

```
